```python
# Filter the data for Rice in Rabi season from Chittoor
filtered_data = df[
    (df['Crop'] == 'Rice') |  # Crop is Rice
    (df['Season'] == 'Rabi') |  # Season is Rabi
    (df['District_Name'] == 'CHITTOOR')  # District is Chittoor
]

# Extract the data for the last 10 years and sort by year
filtered_data_last_10_years = filtered_data.sort_values(by='Crop_Year', ascending=False).head(10)

# Create a dictionary with years as keys and a tuple (area, production) as values
area_production_by_year = dict(zip(filtered_data_last_10_years['Crop_Year'], zip(filtered_data_last_10_years['Area'], filtered_data_last_10_years['Production'])))

# Display the result
print("Area and production data by year (last 10 years):")
print(area_production_by_year)
```

```python
# Filter the data for Kharif season in Anantapur
filtered_data_kharif_anantapur = df[
    (df['Season'] == 'Kharif') | # Season is Kharif
    (df['District_Name'] == 'ANANTAPUR')|  # District is Anantapur
    (df['Crop']=='Coconut')
]

# Sort by production to find the best-performing crops (if production is the criterion)
sorted_data_by_production = filtered_data_kharif_anantapur.sort_values(by='Production', ascending=False)

# Display the crops with the highest production
best_crops_kharif_anantapur = sorted_data_by_production[['Crop', 'Area', 'Production']]
print("Top-performing crops for Kharif season in Anantapur:")
best_crops_kharif_anantapur
```
